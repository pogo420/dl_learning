```python
import torch
import pandas as pd
```

```python
data_raw = pd.read_csv("data/bonus_dataset.csv")
data_raw.head()
```

```python
performance = torch.tensor(data_raw["performance"].values, dtype=torch.float32)
projects_completed = torch.tensor(data_raw["projects_completed"].values, dtype=torch.float32)
bonus = torch.tensor(data_raw["bonus"].values, dtype=torch.float32)
years_of_experience = torch.tensor(data_raw["years_of_experience"].values, dtype=torch.float32)
```

## Batch Gradient Descend

```python
w1 = torch.rand(1, requires_grad=True)
w2 = torch.rand(1, requires_grad=True)
w3 = torch.rand(1, requires_grad=True)
bias = torch.rand(1, requires_grad=True)

epochs = 10000
learning_rate = 0.005
# Batch gradient descent, calculates gradient across the entire data set in each iteration to 
# minimize loss.
for epoch in range(epochs):
    predicted_bonus = w1*performance + w2* years_of_experience + w3* projects_completed + bias
    loss = ((predicted_bonus - bonus)**2).mean()
    loss.backward()
    with torch.no_grad():
        w1 -= (learning_rate*w1.grad)
        w2 -= (learning_rate*w2.grad)
        w3 -= (learning_rate*w3.grad)
        bias -= (learning_rate*bias.grad)
    w1.grad.zero_()
    w2.grad.zero_()
    w3.grad.zero_()
    bias.grad.zero_()

    if epoch % 1000 == 0:
        print(f"Epoch: {epoch}, loss:{loss.item():0.2f}")
```

# Mini batch gradient descend

```python
w1 = torch.rand(1, requires_grad=True)
w2 = torch.rand(1, requires_grad=True)
w3 = torch.rand(1, requires_grad=True)
bias = torch.rand(1, requires_grad=True)

epochs = 10000
learning_rate = 0.005
batch_size  = 10
# Mini Batch gradient descent, calculates gradient across the subset of data set, Values are update in batch
for epoch in range(epochs):
    # batches
    for batch in range(0, len(data_raw), batch_size):
        predicted_bonus = w1*performance[batch: batch+batch_size] + w2* years_of_experience[batch: batch+batch_size] + w3* projects_completed[batch: batch+batch_size] + bias
        loss = ((predicted_bonus - bonus[batch: batch+batch_size])**2).mean()
        loss.backward()
        with torch.no_grad():
            w1 -= (learning_rate*w1.grad)
            w2 -= (learning_rate*w2.grad)
            w3 -= (learning_rate*w3.grad)
            bias -= (learning_rate*bias.grad)
        w1.grad.zero_()
        w2.grad.zero_()
        w3.grad.zero_()
        bias.grad.zero_()

    if epoch % 1000 == 0:
        print(f"Epoch: {epoch}, loss:{loss.item():0.2f}")
```

# Sochastic gradient descend

```python
w1 = torch.rand(1, requires_grad=True)
w2 = torch.rand(1, requires_grad=True)
w3 = torch.rand(1, requires_grad=True)
bias = torch.rand(1, requires_grad=True)

epochs = 10000
learning_rate = 0.005
# Mini Batch gradient descent, calculates gradient across the subset of data set, Values are update in batch
for epoch in range(epochs):
    # batches
    for i in range(0, len(data_raw)):
        predicted_bonus = w1*performance[i] + w2* years_of_experience[i] + w3* projects_completed[i] + bias
        loss = ((predicted_bonus - bonus[i])**2).mean()
        loss.backward()
        with torch.no_grad():
            w1 -= (learning_rate*w1.grad)
            w2 -= (learning_rate*w2.grad)
            w3 -= (learning_rate*w3.grad)
            bias -= (learning_rate*bias.grad)
        w1.grad.zero_()
        w2.grad.zero_()
        w3.grad.zero_()
        bias.grad.zero_()

    if epoch % 1000 == 0:
        print(f"Epoch: {epoch}, loss:{loss.item():0.2f}")
```
