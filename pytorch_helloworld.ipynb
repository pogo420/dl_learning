```python
x = torch.tensor([1.0, 2.0, 3.5],requires_grad=True)
y = torch.tensor([1.9, 2.89, 3.9],requires_grad=True)
z = x*2 + y**3 # Equation 
z.sum().backward() # For backward propagation value must be scalar; To get the grad values  we need to do backward (backward propagetion)
```
